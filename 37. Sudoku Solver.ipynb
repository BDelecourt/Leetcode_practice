```python
"""
Write a program to solve a Sudoku puzzle by filling the empty cells.
A sudoku solution must satisfy all of the following rules:
Each of the digits 1-9 must occur exactly once in each row.
Each of the digits 1-9 must occur exactly once in each column.
Each of the digits 1-9 must occur exactly once in each of the 9 3x3 sub-boxes of the grid.
The '.' character indicates empty cells.

Example 1:
Input: board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
Output: [["5","3","4","6","7","8","9","1","2"],["6","7","2","1","9","5","3","4","8"],["1","9","8","3","4","2","5","6","7"],["8","5","9","7","6","1","4","2","3"],["4","2","6","8","5","3","7","9","1"],["7","1","3","9","2","4","8","5","6"],["9","6","1","5","3","7","2","8","4"],["2","8","7","4","1","9","6","3","5"],["3","4","5","2","8","6","1","7","9"]]
Explanation: The input board is shown above and the only valid solution is shown below:

Constraints:
board.length == 9
board[i].length == 9
board[i][j] is a digit or '.'.
It is guaranteed that the input board has only one solution.
"""
```

```python
import time
from IPython.display import clear_output

def display_sudoku(board,dict_nums,k,l):
    clear_output(wait=True)
    print(f"current index are i={k} and j={l}")
    print(f"previous candidates tested:{dict_nums[(k,l-1)] if (k,l-1) in dict_nums and l>0 else None }" )
    print(f"Potential candidates tested:{dict_nums[(k,l)] if (k,l) in dict_nums else None }" )
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("-" * 21)
        row = ""
        for j in range(9):
            if j % 3 == 0 and j != 0:
                row += "| "
            if board[i][j]=="." and (i,j) in dict_nums:
                row += dict_nums[(i,j)][-1]+ " "
            else:
                row += board[i][j] + " "
        print(row)
    time.sleep(0.1)
```

```python
import random
def solveSudoku(board):
    """
    :type board: List[List[str]]
    :rtype: None Do not return anything, modify board in-place instead.
    """
    def isValidSudoku(board,dict_nums):
        """
        Validates a 9x9 Sudoku board.
        Returns True if the board is valid according to Sudoku rules.
        """
        seen = set()
        
        for i in range(9):
            for j in range(9):
                num = board[i][j]
                if num == "." and (i,j) in dict_nums:
                    num = dict_nums[(i,j)][-1]
                elif num == ".":
                    continue
                row_key = ("row", i, num)
                col_key = ("col", j, num)
                box_key = ("box", i // 3, j // 3, num)

                if row_key in seen or col_key in seen or box_key in seen:
                    return False

                seen.update([row_key, col_key, box_key])

        return True
    
    list_nums=["1","2","3","4","5","6","7","8","9"]
    dict_nums={}
    i,j=0,0
    forward=True
    #display_sudoku(board,dict_nums,i,j)
    while i<9:
        while j<9:
            if board[i][j]==".":
                if (i,j) not in dict_nums:
                    dict_nums[(i,j)]=[random.choice(list_nums)]
                    #display_sudoku(board,dict_nums,i,j)
                elif len(dict_nums[(i,j)])==9:
                    forward=False
                    dict_nums.pop((i,j))
                    #display_sudoku(board,dict_nums,i,j)
                    if j>0:
                        j-=1
                        #display_sudoku(board,dict_nums,i,j)
                    else: 
                        i,j=i-1,8
                        #display_sudoku(board,dict_nums,i,j)
                else:
                    candidate=random.choice(list_nums)
                    while candidate in dict_nums[(i,j)]:
                        candidate=random.choice(list_nums)
                    dict_nums[(i,j)].append(candidate) 
                    #display_sudoku(board,dict_nums,i,j)
                    if isValidSudoku(board,dict_nums):
                        forward=True
                        j+=1
                
            else:
                if forward:
                    j+=1
                else:
                    if j>0:
                        j-=1 
                    else: 
                        i,j=i-1,8
        j=0
        i+=1
    
    for i in range(9):
        for j in range(9):
            if board[i][j]==".":
                board[i][j]=dict_nums[(i,j)][-1]

    return board
```

```python
board = [["5","3",".",".","7",".",".",".","."],
         ["6",".",".","1","9","5",".",".","."],
         [".","9","8",".",".",".",".","6","."],
         ["8",".",".",".","6",".",".",".","3"],
         ["4",".",".","8",".","3",".",".","1"],
         ["7",".",".",".","2",".",".",".","6"],
         [".","6",".",".",".",".","2","8","."],
         [".",".",".","4","1","9",".",".","5"],
         [".",".",".",".","8",".",".","7","9"]]
solveSudoku(board)
```

```python
def solveSudoku_opti(board):
    rows = [set() for _ in range(9)]
    cols = [set() for _ in range(9)]
    boxes = [set() for _ in range(9)]
    empties = []

    # Initialize the sets with existing values and collect empty cells
    for i in range(9):
        for j in range(9):
            num = board[i][j]
            if num == ".":
                empties.append((i, j))
            else:
                rows[i].add(num)
                cols[j].add(num)
                boxes[(i // 3) * 3 + j // 3].add(num)

    def backtrack(index):
        if index == len(empties):
            return True  # All cells are filled successfully

        i, j = empties[index]
        box_id = (i // 3) * 3 + j // 3

        for num in map(str, range(1, 10)):
            if num not in rows[i] and num not in cols[j] and num not in boxes[box_id]:
                # Place the number
                board[i][j] = num
                rows[i].add(num)
                cols[j].add(num)
                boxes[box_id].add(num)

                if backtrack(index + 1):
                    return True

                # Undo the move
                board[i][j] = "."
                rows[i].remove(num)
                cols[j].remove(num)
                boxes[box_id].remove(num)

        return False  # No valid number was found

    backtrack(0)
    return board
```
