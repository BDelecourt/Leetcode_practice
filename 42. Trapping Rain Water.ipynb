```python
"""
Given n non-negative integers representing an elevation map
where the width of each bar is 1, compute how much water it can trap after raining.

Example 1:
Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. 
In this case, 6 units of rain water (blue section) are being trapped.

Example 2:
Input: height = [4,2,0,3,2,5]
Output: 9

Constraints:
n == height.length
1 <= n <= 2 * 104
0 <= height[i] <= 105
"""
```

```python
import matplotlib.pyplot as plt
```

```python
def trap(height):
    """
    :type height: List[int]
    :rtype: int
    """
    def plot_histogram(heights):
        fig, ax = plt.subplots(figsize=(10, 6))
        
        # Create bars
        bar_positions = list(range(len(heights)))
        ax.bar(bar_positions, heights, width=1, color='skyblue', edgecolor='black')

        # Set labels and title
        ax.set_xlabel('Index')
        ax.set_ylabel('Height')
        ax.set_title('Histogram of Heights for Trapping Rain Water')
        ax.set_xticks(bar_positions)
        ax.set_ylim(0, max(heights) + 2)

        # Show grid
        ax.grid(axis='y', linestyle='--', alpha=0.7)

        plt.show()
    
    plot_histogram(height)

    left, right = 0, len(height) - 1
    left_max = right_max = 0
    trapped = 0

    while left < right:
        if height[left] < height[right]:
            left_max = max(left_max, height[left])
            trapped += left_max - height[left]
            left += 1
        else:
            right_max = max(right_max, height[right])
            trapped += right_max - height[right]
            right -= 1

    return trapped

```

```python
height = [0,1,0,2,1,0,1,3,2,1,2,1]
trap(height)
```

```python
height = [4,2,0,3,2,5]
trap(height)
```

```python
height = [5, 4, 1, 2]
trap(height)
```
